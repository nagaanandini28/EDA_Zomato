## Assignment
Find the top 10 cuisines

```python
final_df.columns
```

```python
from collections import Counter

cuisine_counter = Counter()
for cuisine_list in final_df['Cuisines'].dropna():  # Handle missing values
    cuisines = cuisine_list.split(', ')
    cuisine_counter.update(cuisines)

top_10_individual_cuisines = [cuisine[0] for cuisine in cuisine_counter.most_common(10)]
print(top_10_individual_cuisines)
```
